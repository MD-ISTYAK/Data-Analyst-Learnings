```python
import matplotlib.pyplot as plt
#linear graph
x=[1,2,3,4,5]
y=[7,6,8,7,6]
plt.plot(x,y,'r',marker='o')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[7,6,8,7,6]
plt.plot(x,y,'r',marker='o')
plt.plot(y,marker='>')
plt.plot(x,marker='*')
plt.show()
```

```python
#bar plot --vertically
x=[1,2,3,4,5]
y=[7,6,8,7,6]
c=['r','g','b','y','orange']
plt.bar(x,y,color=c, width=0.3)
plt.show()
```

```python
#bar plot --horizontally
x=["A","B","C","D","E"]
y=[7,6,8,7,6]
plt.barh(x,y,color='r', height=0.3)
plt.show()
```

```python
#Scattered Plot
x=[1,2,3,4,5]
y=[7,6,8,7,6]
plt.title("Scatter plot")
plt.xlabel("Month")
plt.ylabel("Years")
c=['r','g','b','y','orange']
plt.scatter(x,y,color=c,s=150,marker='*')
plt.show()
```
