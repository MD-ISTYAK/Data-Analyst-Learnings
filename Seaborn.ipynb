```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var1=[8,3,4,5,6]
var2=[2,3,9,8,4]
#graph using matplotlib
plt.plot(var1,var2)
plt.show()
```

```python
#graph using seaborn library
sns.lineplot(x=var1,y=var1)
plt.show()
```

```python
df=pd.DataFrame({"var1":var1,"var2":var2})
df
```

```python
sns.lineplot(x="var1",y="var2",data=df)
plt.show()
```

```python
df1=sns.load_dataset('penguins')
df1
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=df1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", hue="sex",data=df1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", hue="sex",palette="rocket_r" ,data=df1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", hue="sex",palette="rocket_r" ,style="sex",markers=["o",">"],data=df1)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df1)
plt.show()
```

```python
#histogram  in seaborn
sns.displot(df1["flipper_length_mm"])
plt.show()
```

```python
#using bins in histogram to set the intervals
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
#changing the histogram color
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230],color="red")
plt.show()
```

```python
#using kde=kernel density
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230] , kde=True)
plt.show()
```
